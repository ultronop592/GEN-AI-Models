```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
```

```python
!pip show torch
```

```python
model_id1="CompVis/stable-diffusion-v1-4"
model_id2="stabilityai/stable-diffusion-2-1"

pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16)
pipe.to("cuda")
```

```python
prompt ="""A futuristic city at night with glowing neon lights, flying cars, and tall skyscrapers under a starry sky."""
```

```python
image = pipe(prompt).images[0]
```

```python
image
```

```python
print("[PROMPT]:",prompt)
plt.imshow(image)
plt.axis("off")
plt.show()
```
